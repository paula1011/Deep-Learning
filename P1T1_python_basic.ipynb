##Ejercicio 1: productos + IVA

```python
prod1= input("input 1 product: ")
prod2= input("input 2 product: ")
prod3= input("input 3 product: ")
prod4= input("input 4 product: ")
prod5= input("input 5 product: ")
suma= (int(prod1)+int(prod2)+int(prod3)+int(prod4)+int(prod5))
iva= suma*0.19
total= suma+ iva
print("valor bruto:",suma,"\nIVA:",iva,"\nTOTAL:",total )
```
